Análise da "Gestão Júlio Casares" em dados

Para a análise de desempenho do São Paulo nas copas, vou atribuir a cada fase de mata‑mata um valor numérico correspondente ao “top” em que o time terminou. Por exemplo, quando o São Paulo chegar às quartas de final, o valor será 8, indicando que a equipe ficou entre os oito melhores (top 8).

Desempenho Paulista

```r
library(ggplot2)

X <- c(1, 2, 8, 8, 4)
anos <- 2021:2025
df_paulista <- data.frame(ano = anos, top = X)

ggplot(df_paulista, aes(x = ano, y = top)) +
  geom_line(color = "#cc0000", linewidth = 1) +
  geom_point(color = "#cc0000", size = 3) +
  scale_y_reverse(limits = c(8, 1), breaks = c(1, 2, 4, 8)) +
  labs(
    x = "Ano",
    y = "Posição (top na competição)",
    title = "Desempenho do São Paulo no Campeonato Paulista (2021–2025)"
  ) +
  theme_minimal()

```

Análise dos números da gestão casares


```r
anos       <- 2021:2025
gols_pro   <- c(101, 131, 105, 106, 83)
gols_contra<- c(68,  90,  72,  74, 79)
vitorias   <- c(29,  36,  33,  32, 28)
derrotas   <- c(17,  18,  21,  19, 22)
empates    <- c(24,  23,  17,  17, 16)
jogos      <- c(70,  77,  71,  68, 66)

df_temp <- data.frame(
  ano        = anos,
  gols_pro   = gols_pro,
  gols_contra= gols_contra,
  vitorias   = vitorias,
  derrotas   = derrotas,
  empates    = empates,
  jogos      = jogos
)
```

```r
df_temp$saldo_gols     <- df_temp$gols_pro - df_temp$gols_contra
df_temp$gols_pro_jogo  <- df_temp$gols_pro    / df_temp$jogos
df_temp$gols_con_jogo  <- df_temp$gols_contra / df_temp$jogos
df_temp$saldo_por_jogo <- df_temp$saldo_gols  / df_temp$jogos

df_temp$pontos         <- 3*df_temp$vitorias + df_temp$empates
df_temp$aproveitamento <- df_temp$pontos / (3*df_temp$jogos)
```

```r
library(ggplot2)

ggplot(df_temp, aes(x = ano)) +
  geom_line(aes(y = gols_pro_jogo, color = "Gols pró/jogo"), linewidth = 1) +
  geom_line(aes(y = gols_con_jogo, color = "Gols contra/jogo"), linewidth = 1) +
  labs(x = "Ano", y = "Gols por jogo",
       title = "Média de gols por jogo do São Paulo (todas as competições)") +
  scale_color_manual(values = c("Gols pró/jogo" = "#cc0000",
                                "Gols contra/jogo" = "#333333"),
                     name = NULL) +
  theme_minimal()

ggplot(df_temp, aes(x = ano, y = aproveitamento)) +
  geom_line(color = "#cc0000", linewidth = 1) +
  geom_point(color = "#cc0000", size = 3) +
  scale_y_continuous(labels = scales::percent_format(accuracy = 1)) +
  labs(x = "Ano", y = "Aproveitamento na temporada",
       title = "Aproveitamento geral do São Paulo na temporada (2021–2025)") +
  theme_minimal()
```

```r
df_temp$prop_vit   <- df_temp$vitorias / df_temp$jogos
df_temp$prop_emp   <- df_temp$empates  / df_temp$jogos
df_temp$prop_der   <- df_temp$derrotas / df_temp$jogos
```

```r
library(tidyr)

df_prop <- df_temp[, c("ano", "prop_vit", "prop_emp", "prop_der")] |>
  pivot_longer(
    cols = c("prop_vit", "prop_emp", "prop_der"),
    names_to = "resultado",
    values_to = "proporcao"
  )
```

```r
df_prop$resultado <- factor(
  df_prop$resultado,
  levels = c("prop_vit", "prop_emp", "prop_der"),
  labels = c("Vitória", "Empate", "Derrota")
)

ggplot(df_prop, aes(x = factor(ano), y = proporcao, fill = resultado)) +
  geom_col() +
  scale_y_continuous(labels = scales::percent_format(accuracy = 1)) +
  scale_fill_manual(values = c("Vitória" = "#228B22",
                               "Empate"  = "#CCCC00",
                               "Derrota" = "#CC0000"),
                    name = "Resultado") +
  labs(
    x = "Ano",
    y = "Proporção de resultados",
    title = "Perfil de temporada do São Paulo (vitórias, empates e derrotas)"
  ) +
  theme_minimal()
```
